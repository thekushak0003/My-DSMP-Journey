#### Selecting rows from a DataFrame

- **iloc** - searches using index positions
- **loc** - searches using index labels 

```python
# single row
ipl.iloc[3]
```

```python
movies.iloc[1:12:2,4:9:2] #.iloc[rows,column]
```

```python
movies.loc[0:10:2,'title_y':'runtime':2] #.loc[rows,column]
```

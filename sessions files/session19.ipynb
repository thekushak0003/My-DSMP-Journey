#### GroupBy

```python
import pandas as pd 
import numpy as np 
```

```python
movies=pd.read_csv('imdb-top-1000.csv')
movies.head(10)
```

```python
genre=movies.groupby('Genre')
```

```python
genre.sum()
```

```python
# Find the top 3 genres by total earning
movies.groupby('Genre').sum()['Gross'][:3]
movies.groupby('Genre')['Gross'].sum()[:3] # This method is more efficient
```

```python
# find the genre with highest avg IMDB rating
movies.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).head(1)
```

```python
# find director with most popularity
movies.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head(1)
```

```python
# find the highest rated movie of each genre
movies.groupby('Genre')['IMDB_Rating'].max().sort_values(ascending=False)
```

```python
# find number of movies done by each actor
movies.groupby('Star1')['Series_Title'].count().sort_values(ascending=False)
```

#### Group attributes and Methods

```python
# find total number of groups -> len
len(movies.groupby('Genre'))
```

```python
# find items in each group -> size
movies.groupby('Genre').size()
```

```python
# first()/last() -> nth item
genres = movies.groupby('Genre')
# genres.first()
# genres.last()
genres.nth(6) # evry grp's 7th movie
```

```python
# get_group -> vs filtering
genre.get_group('Drama')#this is much more faster than traditional filtering ways
```

```python
genres.groups
```

```python
genres.sample(2,replace=True)
```

```python
genres.nunique()
```

#### agg_methods

```python
# Passing dict
genres.agg(
    {
        'Runtime':'mean',
        'IMDB_Rating':'mean',
        'No_of_Votes':'sum',
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

```python
genres.agg({
	'Runtime': ['min', 'max', 'mean', 'sum'],
	'IMDB_Rating': ['min', 'max', 'mean'],
	'No_of_Votes': ['min', 'max'],
	'Gross': ['min'],
	'Metascore': 'max'
})
```

#### looping on groups

```python
df = pd.DataFrame(columns=movies.columns)
# Collect all DataFrames in a list and concatenate at the end
dfs = []
for group, data in genres:
    dfs.append(data[data['IMDB_Rating']==data['IMDB_Rating'].max()])
df=pd.concat(dfs, ignore_index=True)
df
```

#### Split --> Apply --> Combine strategy

```python
def fnameA(group):
    return group['Series_Title'].str.startswith('A').sum()
```

```python
genres.apply(fnameA, include_groups=False)
```

```python
# find ranking of each movie in the group according to IMDB score
def rank_movie_imdb(group):
    group['IMDB_Rating'].rank(ascending=False)
    return group
```

```python
genres.apply(rank_movie_imdb,include_groups=False)
```

```python
# find normalized IMDB rating group wise
def normal(group):
    group['norm_rating'] = (group['IMDB_Rating']-group['IMDB_Rating'].min())/(group['IMDB_Rating'].max()-group['IMDB_Rating'].min())
    return group
genres.apply(normal,include_groups=False)
```

#### Multiple groupby

```python
duo = movies.groupby(['Director','Star1'])
duo
# size
duo.size()
# get_group
duo.get_group(('Aamir Khan','Amole Gupte'))
```

```python
# Find the most earning Actor-Director duo
duo['Gross'].sum().sort_values(ascending=False)
```

```python
# find the best(in-terms of metascore(avg)) actor->genre combo
movies.groupby(['Star1','Genre'])['Metascore'].mean().reset_index().sort_values('Metascore',ascending=False).head(10)
```

```python
# agg on multiple groupby
duo.agg({
	'Runtime': ['min', 'max', 'mean'],
	'IMDB_Rating': ['min', 'max', 'mean'],
	'No_of_Votes': ['min', 'max', 'mean'],
	'Gross': ['min', 'max', 'mean'],
	'Metascore': ['min', 'max', 'mean']
})
```

```python
ipl = pd.read_csv('deliveries.csv')
ipl.sample(10)
```

```python
# find the top 10 batsman in terms of runs
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# find the batsman with max no of sixes
six=ipl[ipl['batsman_runs']==6]
six.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(10)
```

```python
# find batsmans with most number of 4's and 6's in last 5 overs
temp_df1 = ipl[ipl['over'] > 15]
temp_df1 = temp_df1[(temp_df1['batsman_runs'] == 4) | (temp_df1['batsman_runs'] == 6)]
temp_df1.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(10)
```

```python
# Find SK Raina's record against all teams
temp_df2 = ipl[ipl['batsman'] == 'SK Raina']
temp_df2.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False)
```

```python
def highest(batsman):
  temp_df = ipl[ipl['batsman'] == batsman]
  return temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

### 2-D Line Plot

```python
# import the library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('fivethirtyeight')
```

```python
# plotting a simple function
price = [48000,54000,57000,50000,49000,45000,44000,48000,42000,40000,46000]
year = [2015,2016,2017,2018,2019,2020,2021,2022,2023,2024,2025]
plt.plot(year,price)
```

```python
# from a pandas dataframe
batsman = pd.read_csv('sharma-kohli.csv')
batsman
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
# plotting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# labels title
plt.plot(batsman['index'],batsman['V Kohli'],color='orange')
plt.plot(batsman['index'],batsman['RG Sharma'],color='green')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('IPL Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='orange',linestyle='solid',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='dashdot',linewidth=2)

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('IPL Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='orange',linestyle='solid',linewidth=3,marker='o',label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='dashdot',linewidth=2,marker='D',label='Rohit')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('IPL Season')
plt.ylabel('Runs Scored')
plt.legend(loc='best')
```

```python
### Handling outliers
price = [48000,54000,57000,50000,49000,45000,44000,480000,42000,40000,46000]
year = [2015,2016,2017,2018,2019,2020,2021,2022,2023,2024,2025]
plt.plot(year,price)
plt.ylim(0,75000)
```

### Scatter Plot

```python
# plt.scatter simple function
x = np.linspace(-10,10,50)
y = 10*x + 3 + np.random.randint(0,300,50)
plt.scatter(x,y)
```

```python
# plt.scatter on pandas data
df = pd.read_csv('batter.csv')
df = df.head(100)
plt.scatter(df['avg'],df['strike_rate'],color='red',marker='*')
plt.title('Avg and SR analysis of Top 100 Batsman')
plt.xlabel('Average')
plt.ylabel('SR')
```

```python
tips = sns.load_dataset('tips')
plt.plot(tips['total_bill'],tips['tip'],'*')
```

### Bar Chart

```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['Red','Blue','Green','Yellow','Pink']

plt.bar(colors,children,color='orange')
```

```python
df = pd.read_csv('batsman_season_record.csv')
plt.bar(np.arange(df.shape[0]) - 0.25,df['2015'],width=0.2,color='orange')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='blue')
plt.bar(np.arange(df.shape[0]) + 0.25,df['2017'],width=0.2,color='green')
plt.xticks(np.arange(df.shape[0]), df['batsman'])
plt.xticks(rotation=30)
plt.show()
```

```python
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')

plt.legend()
plt.show()
```

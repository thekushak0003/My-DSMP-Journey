```python
import seaborn as sns
```

```python
# import datasets
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

### Categorical Plots

#### Categorical Scatter Plot

- Stripplot
- Swarmplot

#### Categorical Distribution Plots

- Boxplot
- Violinplot

#### Categorical Estimate Plot -> for central tendency

- Barplot
- Pointplot
- Countplot

#### Figure level function -> `catplot`

#### Stripplot

```python
# stripplot
# axes level function
sns.stripplot(data=tips,x='day',y='total_bill',hue='sex')
```

#### Catplot

```python
# using catplot
# figure level function
sns.catplot(data=tips, x='day',y='total_bill',kind='strip',hue='sex')
```

#### Jitter

```python
sns.catplot(data=tips, x='day',y='total_bill',kind='strip',jitter=0.25,hue='sex')
```

#### Swarmplot

```python
# swarmplot
sns.catplot(data=tips, x='day',y='total_bill',kind='swarm')
```

```python
sns.swarmplot(data=tips, x='day',y='total_bill')
```

```python
# hue
sns.swarmplot(data=tips, x='day',y='total_bill',hue='sex')
```

#### Boxplot

```python
# Box plot
sns.boxplot(data=tips,x='day',y='total_bill')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='box',hue='sex')
```

```python
sns.catplot(data=tips,x='total_bill',kind='box')
```

#### Voilin plot = (Boxplot + KDEplot)

```python
sns.violinplot(data=tips,x='day',y='total_bill')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='violin',hue='sex',split=True)
```

####  Bar Plot

```python
import numpy as np
sns.barplot(data=tips,x='sex',y='total_bill',hue='smoker',estimator=np.min)
```

```python
sns.barplot(data=tips, x='sex', y='total_bill',errorbar=None)
```

```python
# point plot
sns.pointplot(data=tips,x='sex',y='total_bill',hue='smoker',errorbar=None)
```

```python
# countplot
sns.countplot(data=tips,x='day',hue='sex')
```

```python
# faceting using catplot
sns.catplot(data=tips,x='sex',y='total_bill',col='time',kind='box',row='smoker')
```

### Regression Plots

- regplot
- lmplot

In the simplest invocation, both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.

#### Regplot

```python
# axes level
# hue parameter is not available
sns.regplot(data=tips,x='total_bill',y='tip')
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex')
```

```python
# residplot
sns.residplot(data=tips,x='total_bill',y='tip')
```

### Multigrid plot

#### A second way to plot Facet plots -> FacetGrid

```python
# figure level -> relplot -> displot -> catplot -> lmplot
sns.catplot(data=tips,x='sex',y='total_bill',kind='violin',col='day',row='time')
```

```python
g=sns.FacetGrid(data=tips,col='day',row='time')
g.map(sns.violinplot,'sex','total_bill',order=['Male','Female'])
```

#### Plotting Pairwise Relationship (PairGrid Vs Pairplot)

```python
sns.pairplot(iris,hue='species')
```

```python
# pair grid
g = sns.PairGrid(data=iris,hue='species')
# g.map
g.map(sns.scatterplot)
```

```python
# map_diag -> map_offdiag
g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.boxplot)
g.map_offdiag(sns.kdeplot)
```

```python
# map_diag -> map_upper -> map_lower
g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
# vars
g = sns.PairGrid(data=iris,hue='species',vars=['sepal_width','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

### JointGrid Vs Jointplot

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='scatter',hue='sex')
```

```python
g = sns.JointGrid(data=tips,x='total_bill',y='tip',hue='sex')
g.plot(sns.kdeplot,sns.violinplot)
```
